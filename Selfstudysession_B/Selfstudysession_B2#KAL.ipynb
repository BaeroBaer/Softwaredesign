## Exercise B.2 (Lists, conditionals and loops) For two given lists list1 and list2 of integers with equal length define the following new lists:

    list_zip which combines the two lists to a single list like a zip. So start with the first element of list1, followed by the first element of list2, than the second elements until all elements of the lists are combined.
    list_odd which contains only the odd integers of list1, followed by the odd integers of list2.
    list_zip_reverse which combines the elements of the lists like for list_zip but starts at the last element of list1.

Example:

For list1 = [1, 2, 3, 4, 5] and list2 = [11, 12, 13, 14, 15], we get

    list_zip = [1, 11, 2, 12, 3, 13, 4, 14, 5, 15]
    list_odd = [1, 3, 5, 11, 13, 15]
    list_zip_reverse = [5, 15, 4, 14, 3, 13, 2, 12, 1, 11]

```python
## Variante A

list1=[1,2,3,4,5]
list2=[11,12,13,14,15]
list_zipped=[]
list_odd=[]
list_zip_reversed=[]


## schleife fuer zipped
for i in range(len(list1)):
    list_zipped.append(list1[i])
    list_zipped.append(list2[i])
    
print(list_zipped)
## schleife fuer odd
for i in [*list1, *list2]:
    if i % 2 != 0:
        list_odd.append(i)
    #if list2[i] % 2 != 0:
    #    list_odd.append(list2[i])

print(list_odd)

## schleife fuer reversed zipped
for i in range(len(list1) - 1, -1, -1) : 
    list_zip_reversed.append(list1[i])
    list_zip_reversed.append(list2[i])

print(list_zip_reversed)  
```

```python
# Exercise B.3 (Fizz Buzz)
# Schreibe ein Programm, das die Zahlen von 1 bis 100 ausgibt.
# - durch 3 teilbar -> Fizz
# - durch 5 teilbar -> Buzz
# - durch 3 und 5 teilbar -> FizzBuzz

for i in range(1, 101):
    out = ""
    if i % 3 == 0:
        out += "Fizz"
    if i % 5 == 0:
        out += "Buzz"
    print(out if out else i)
```

```python
##Exercise B.4
#Exercise B.4 (Difference between assignment, copy and deep copy) Have a look at the module copy and the functions copy, deepcopy as well as the standard assignment operator =.

#Run the following code and explain its output:


import copy

list1 = [[1, 2, 3], [4, 5, 6], ["a", "b", "c"], 2]
list2 = list1     #
list3 = copy.copy(list1)      #copy flache Kopie äußere Objekt ist neu aber das innere Objekt ist das gleiche wie in der alten Liste. Änderungen an der neuen Liste können Auswirkungen auf die alte Liste haben, wenn es sich um verschachtelte Objekte handelt, da sie auf die gleichen inneren Objekte verweisen.
list4 = copy.deepcopy(list1)  #deep copy erstellt eine komplett neue Liste, die von der alten Liste unabhängig ist. Alle Elemente werden rekursiv kopiert, so dass Änderungen an der neuen Liste keine Auswirkungen auf die alte Liste haben.

print("List # \tID\tEntries")
print("1\t", id(list1), "\t", list1)
print("2\t", id(list2), "\t", list2)
print("3\t", id(list3), "\t", list3)
print("4\t", id(list4), "\t", list4)

list1[3] = -2

print("List # \tID\tEntries")
print("1\t", id(list1), "\t", list1)
print("2\t", id(list2), "\t", list2)
print("3\t", id(list3), "\t", list3)
print("4\t", id(list4), "\t", list4)

list2[2][2] = 9

print("List # \tID\tEntries")
print("1\t", id(list1), "\t", list1)
print("2\t", id(list2), "\t", list2)
print("3\t", id(list3), "\t", list3)
print("4\t", id(list4), "\t", list4)

list1.append([0, 8, 15])

print("List # \tID\tEntries")
print("1\t", id(list1), "\t", list1)
print("2\t", id(list2), "\t", list2)
print("3\t", id(list3), "\t", list3)
print("4\t", id(list4), "\t", list4)
```

erste Ausgabe: ID von liste1 und liste2 sind identisch, da liste2 = liste1
liste3 hat andere ID (äußere Liste neu), innere Objekte werden geteilt
liste4 hat eigene IDs (deepcopy) — komplett unabhängig

nach list1[3] = -2: nur liste1 und liste2 zeigen -2; liste3 und liste4 behalten 2
nach list2[2][2] = 9: liste1, liste2 und liste3 zeigen 9 (inneres Objekt geteilt); liste4 bleibt unverändert

```python
##B.5 (Monte Carlo Simulation) Write a program that estimates the value of pi using a Monte Carlo simulation. The idea is to randomly generate points in a unit square and count how many fall inside a quarter circle inscribed within that square. The ratio of points inside the quarter circle to the total number of points can be used to estimate pi.
import random
import math

# zähle Punkte im Viertelkreis
def in_unit_circle(N):
    M = 0
    for _ in range(N):
        x = random.random()
        y = random.random()
        if x*x + y*y <= 1:
            M += 1
    return M

# schätze pi: pi ≈ 4 * M / N
def estimate_pi(N):
    M = in_unit_circle(N)
    return 4 * M / N

# fehler: |geschätzt - tatsächlich|
def get_accuracy(N):
    return abs(estimate_pi(N) - math.pi)

# test mit verschiedenen N — größere N geben bessere Genauigkeit
print(f"Exact pi = {math.pi:.6f}\n")
for N in (1000, 10000, 100000, 1000000):
    pi_hat = estimate_pi(N)
    error = get_accuracy(N)
    print(f"N={N:>7}  pi≈{pi_hat:.6f}  error={error:.6f}")
```

```python
## Exercise B.8 (Functions can be arguments)

import random
from typing import Callable

def in_unit_circle(N: int) -> int:
    M = 0
    for _ in range(N):
        x = random.random()
        y = random.random()
        if x * x + y * y <= 1:
            M += 1
    return M

def estimate_pi(N: int, counter_fn: Callable[[int], int]) -> float:
    M = counter_fn(N)
    return 4 * M / N

# Test
for N in (1_000, 10_000, 100_000):
    print(f"N={N:>7}  pi~{estimate_pi(N, in_unit_circle):.6f}")
```

```python
## Exercise B.9 (Recursion with Fibonacci)

def fibonacci(n: int) -> int:
    if n < 0:
        raise ValueError("n must be >= 0")
    if n == 0:
        return 0
    if n == 1:
        return 1
    return fibonacci(n - 1) + fibonacci(n - 2)

# Test: first 15 values
for i in range(15):
    print(f"fibonacci({i}) = {fibonacci(i)}")
```

```python
# B.10 Working with dictionaries

roman = {
    "I": 1, "IV": 4, "V": 5, "IX": 9,
    "X": 10, "XL": 40, "L": 50, "XC": 90,
    "C": 100, "CD": 400, "D": 500, "CM": 900,
    "M": 1000
}

def roman_to_arabic(number):
    i = 0
    total = 0

    while i < len(number):
        if i + 1 < len(number) and number[i:i+2] in roman:
            total += roman[number[i:i+2]]
            i += 2
        else:
            total += roman[number[i]]
            i += 1

    return total

print(roman_to_arabic("MMMDCCXXIV"))  # 3724
print(roman_to_arabic("MXCIV"))       # 1094
print(roman_to_arabic("VIII"))        # 8
```

```python

##Exercise B.12 (Set, List and Dictionary Comprehensions)  


# 1) flatten mit list comprehension
list_of_sets = [(1, 2, 3), (4, 5, 6), (7, 8)]
flat = [x for group in list_of_sets for x in group]
print(flat)

# 2) (9) ist kein Tupel, sondern int -> würde Fehler geben
# list_of_sets = [(1, 2, 3), (4, 5, 6), (7, 8), (9)]
# flat = [x for group in list_of_sets for x in group]  # TypeError

# 3) (9,) ist 1er-Tupel -> funktioniert
list_of_sets = [(1, 2, 3), (4, 5, 6), (7, 8), (9,)]
flat = [x for group in list_of_sets for x in group]
print(flat)

# 4) mit Listen geht es genauso
list_of_lists = [[1, 2, 3], [4, 5, 6], [7, 8], [9]]
flat = [x for group in list_of_lists for x in group]
print(flat)

# 5) Satz in Wörter teilen
sentence = "I spy with my little eye a tricky list and dict comprehension coming up"
words = sentence.split()
print(words)

# 6) Hint mit map: eindeutige Wortlängen
lengths = set(map(len, words))
print(lengths)

# 7) dict: key=länge, value=set von wörtern
by_length = {}
for w in words:
    l = len(w)
    if l not in by_length:
        by_length[l] = set()
    by_length[l].add(w)
print(by_length)

# 8) alles in einer dictionary comprehension
by_length_comp = {l: {w for w in words if len(w) == l} for l in set(map(len, words))}
print(by_length_comp)
```
